````python
query = f"""
WITH matches AS (
    SELECT
        hashed_identifier
      , to_date(from_unixtime(CAST(order_id_timestamp_min AS INT))) AS rec_date
      , CASE WHEN length(order_id) = 33 THEN '33'
             WHEN length(order_id) = 38 THEN '38'
             ELSE NULL END AS id_type
      , COUNT(*) AS num_matches
    FROM
        {fb_sync}
    WHERE
        length(order_id) > 10
    GROUP BY
        hashed_identifier
      , to_date(from_unixtime(CAST(order_id_timestamp_min AS INT)))
      , CASE WHEN length(order_id) = 33 THEN '33'
             WHEN length(order_id) = 38 THEN '38'
             ELSE NULL END
    )
SELECT
    rec_date
  , id_type
  , SUM(num_matches) AS num_matches
  , COUNT(DISTINCT hashed_identifier) AS num_hashes
  , COUNT(*) AS num_rows
FROM
    matches
GROUP BY
    rec_date
  , id_type
ORDER BY
    rec_date
  , id_type
;"""
print(query)
mapping_counts = hive_query(query, config=config, parse_dates=['rec_date'])
mapping_counts.info()
````

````python
query = f"""
WITH matches AS (
    SELECT
        hashed_identifier
      , to_date(from_unixtime(CAST(order_id_timestamp_min AS INT))) AS rec_date
      , MAX(CASE WHEN length(order_id) = 33 THEN 1 ELSE 0 END) AS has_33
      , MAX(CASE WHEN length(order_id) = 38 THEN 1 ELSE 0 END) AS has_38
      , COUNT(*) AS num_matches
    FROM
        {fb_sync}
    WHERE
        length(order_id) > 10
    GROUP BY
        hashed_identifier
      , to_date(from_unixtime(CAST(order_id_timestamp_min AS INT)))
    )
SELECT
    rec_date
  , SUM(has_33) AS num_33
  , SUM(has_38) AS num_38
  , SUM(has_33 * has_38) AS num_both
  , SUM(num_matches) AS num_matches
  , COUNT(DISTINCT hashed_identifier) AS num_hashes
  , COUNT(*) AS num_rows
FROM
    matches
GROUP BY
    rec_date
ORDER BY
    rec_date
;"""
print(query)
mapping_overlap = hive_query(query, config=config, parse_dates=['rec_date'])
mapping_overlap.info()
````

````python
lookback = 30
start_date = '2018-07-29'
end_date = '2018-09-29'

query = f"""
WITH merged AS (
    SELECT
        a.actuuid,
        a.act_time,
        a.userid,
        a.activity_type,
        MAX(
            CASE WHEN e.evt_type_class = 'IM' AND upper(e.evt_source) = 'FACEBOOK' AND
                      unix_timestamp(e.evt_time) <= unix_timestamp(a.act_time) AND
                      unix_timestamp(e.evt_time) >= unix_timestamp(a.act_time) - {lookback}*24*60*60
                 THEN 1
                 ELSE 0
                 END
        ) AS has_fb_im,
        MAX(
            CASE WHEN e.evt_type_class = 'CL' AND upper(e.evt_source) = 'FACEBOOK' AND
                      unix_timestamp(e.evt_time) <= unix_timestamp(a.act_time) AND
                      unix_timestamp(e.evt_time) >= unix_timestamp(a.act_time) - {lookback}*24*60*60
                 THEN 1
                 ELSE 0
                 END
        ) AS has_fb_cl,
        MAX(
            CASE WHEN upper(e.evt_source) = 'FACEBOOK' AND
                      unix_timestamp(e.evt_time) <= unix_timestamp(a.act_time) AND
                      unix_timestamp(e.evt_time) >= unix_timestamp(a.act_time) - {lookback}*24*60*60
                 THEN 1
                 ELSE 0
                 END
        ) AS has_fb
    FROM
        {albis} a
    LEFT JOIN
        {elbis} e
    ON
        a.userid = e.userid
    WHERE
        TO_DATE(a.act_time) BETWEEN '{start_date}' AND '{end_date}'
    GROUP BY
        a.actuuid,
        a.act_time,
        a.userid,
        a.activity_type
)
SELECT
    activity_type
  , COUNT(*) AS num_rows
  , COUNT(DISTINCT actuuid) AS num_act
  , COUNT(DISTINCT userid) AS num_id
  , SUM(has_fb_im) AS num_fb_im
  , SUM(has_fb_cl) AS num_fb_cl
  , SUM(has_fb) AS num_fb
FROM
    merged
GROUP BY
    activity_type
;"""
hive_query(query, config=config)
````

## Uploading data through the Conversions API

**Step #1: Create an empty Offline Event Set.**

```bash
curl \
-F 'access_token=<ACCESS_TOKEN>' \
-F 'name=Pier 1 Cohort Test Conversions' \
-F 'description=Offline Event Set for Pier 1 cohort tests' \
https://graph.facebook.com/v3.1/1386462108046405/offline_conversion_data_sets
```

**Step #2: Associate the Offline Event Set with the corresponding Dummy Ad Account**. 
```bash
curl -X POST \
-F 'access_token=<ACCESS_TOKEN>' \
-F 'business=1386462108046405' \
-F 'account_id=1452635804762368' \
https://graph.facebook.com/v3.1/<OFFLINE_EVENT_SET_ID>/adaccounts
```

**Step #3: Reach out to Facebook Engineering, with the Offline Event Set ID**

**Step #4: Populate the Offline Event Set**

````python
def gen_bb_api(db, table, backbone, start_date, end_date):
    """Generates the stack backbone complying to the Conversions API proposed by Chris"""
    print("Starting gen_bb_sync")
    #start_date = '2018-09-30'  # Sunday
    #end_date = '2018-10-13'  # Saturday
    query = f"""
    set hive.execution.engine=mr;

    DROP TABLE IF EXISTS {db}.{table};
    CREATE TABLE IF NOT EXISTS {db}.{table} STORED AS ORC AS
    SELECT
        t1.userid AS match_keys_extern_id
      , t1.ref_timestamp AS event_time
      , t1.actuuid AS actuuid
      , COALESCE(concat_ws('|', COLLECT_SET(t3.order_id)), '') AS custom_data_cookie_ids
      , COALESCE(concat_ws('|', COLLECT_SET(t4.hash_ekey)), '') AS custom_data_ext_ids
      , CASE WHEN t1.response = 1 THEN 'conversion' ELSE 'non-conversion' END AS custom_data_type
    FROM
        {backbone} t1  -- This is the Neustar stack
    LEFT JOIN
        {id_mapping} t2  -- This is an ID mapping that is available for every client
    ON
        t1.userid = t2.userid
    LEFT JOIN
        {fb_sync} t3  -- This comes from FB - Mapping between mcvisid to hashed FB ID
    ON
        t2.hashed_identifier = t3.hashed_identifier
    LEFT JOIN
        {offline_id_map} t4  -- From Chandrika
    ON
        t1.userid = t4.hash_pid
    WHERE
        to_date(from_unixtime(t1.ref_timestamp)) >= '{start_date}' AND
        to_date(from_unixtime(t1.ref_timestamp)) <= '{end_date}'
    GROUP BY
        t1.userid
      , t1.ref_timestamp
      , t1.actuuid
      , CASE WHEN t1.response = 1 THEN 'conversion' ELSE 'non-conversion' END
    ;"""
    print(query)
    hive_query(query, config=config, print_stderr=True)
    
    query = f"""
    SELECT
        COUNT(*) AS num_rows
    FROM
        {db}.{table}
    """
    count = hive_query(query, config=config, print_stderr=True)
    print(f"Num rows in {table}:")
    display(count)

    return f"{db}.{table}"
````

````python
def create_week_boundaries(start, end, first='Sun'):
    """Generates a list of tuples containing the start and end of weeks (Sun-Sat)"""
    start_dt = datetime.datetime.strptime(start, '%Y-%m-%d')
    start_iso = start_dt.isocalendar()
    #start_week = start_iso[0]*100 + start_iso[1]
    end_dt = datetime.datetime.strptime(end, '%Y-%m-%d')
    end_iso = end_dt.isocalendar()
    #end_week = end_iso[0]*100 + end_iso[1]
    if start_dt >= end_dt:
        raise ValueError("Start date must be earlier than end date")
    if start_iso[2] != 7:
        raise ValueError("Start date must be a Sunday")
    if end_iso[2] != 6:
        raise ValueError("End date must be a Saturday")
    max_dt = start_dt
    weeks = []
    counter = 0
    day = datetime.timedelta(days=1)
    while max_dt <= end_dt:
        weeks.append(
            ((start_dt + counter*7*day), (start_dt + ((counter+1)*7 - 1)*day))
        )
        counter += 1
        max_dt = start_dt + ((counter+1)*7 - 1)*day
    return weeks
````
